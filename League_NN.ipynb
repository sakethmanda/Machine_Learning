```python
#We need to split the data set into training and testing sets, before we do anything further.

from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(data, test_size=0.2)
```

```python
def split_xy(data):
    X = data.loc[:, data.columns != 'blueWins'].values
    y = data['blueWins'].values
    return X, y
```

```python
X_train, y_train = split_xy(train_set)
X_test, y_test = split_xy(test_set)

print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

```python
# As a baseline, we use random forest
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier

# Use StandardScaler to normalize data
pipeline = []
pipeline.append(("scaler", StandardScaler()))
pipeline.append(("rf", RandomForestClassifier()))

rf_clf = Pipeline(pipeline)
rf_clf.fit(X_train, y_train)
```

```python
# Evaluate performance
from sklearn.metrics import classification_report
y_pred = rf_clf.predict(X_test)
print(f"Classification report for classifier {rf_clf}:\n"
      f"{classification_report(y_test, y_pred)}\n")
```

```python
from sklearn.neural_network import MLPClassifier

nn_clf = Pipeline([
    ("scaler", StandardScaler()),
    ("lr", MLPClassifier(hidden_layer_sizes=(10,4), solver='sgd'))
])
nn_clf.fit(X_train, y_train)
print(f"Classification report for classifier {nn_clf}:\n"
      f"{classification_report(y_test, nn_clf.predict(X_test))}\n")
```
